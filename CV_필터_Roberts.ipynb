```python
img = cv2.imread('20141006175.jpg')
plt.imshow(img)
```

```python
resize_img = cv2.resize(img, (576,360))
cv2_imshow(resize_img)
```

```python
gray = cv2.cvtColor(resize_img, cv2.COLOR_BGR2GRAY)
#plt.imshow(gray, cmap='gray')
cv2_imshow(gray)
```

```python
height = gray.shape[0]
width = gray.shape[1]

img_01 = np.zeros((height, width))
img_02 = np.zeros((height, width))
img_03 = np.zeros((height, width))

print(gray.shape, img_01.shape)
#plt.imshow(img_01, cmap='gray')
#cv2_imshow(img_01)
```

```python
h_filter = np.array(
            [[1, 0],
            [0, -1]]
            )
v_filter = np.array(
            [[0, 1],
            [-1, 0]]
            )
for y in range(height):
    for x in range(width):
        if y<height-1:
            if x<width-1:
                h_val = np.sum(h_filter*gray[y:y+2,x:x+2])
                img_01[y,x] = h_val
                v_val = np.sum(v_filter*gray[y:y+2,x:x+2])
                img_02[y,x] = v_val
                x_val = np.sqrt(h_val*h_val + v_val*v_val)
                #if x_val<100:
                #    x_val = 0
                #if x_val>255:
                #   x_val = 255
                img_03[y,x] = x_val
```

```python
#plt.imshow(img_01, cmap='gray')
cv2_imshow(img_01)
```

```python
#plt.imshow(img_02, cmap='gray')
cv2_imshow(img_02)
```
